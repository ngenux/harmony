```python
import random

def choose_first_player(A,B):
    """
    This function returns one player by random choosing as an output
    """
    first_player = random.choice([A,B])
    return (first_player)

# for i in range(5):
#     print(choose_first_player('p1','p2'))

# Define an empty cell
empty_cell = '| |'

def init_state_board():
    """
    Define initial state board
    """
    board = [[empty_cell for x in range(3)] for y in range(3)]
    turn = 0
    return board,turn

def format_board(player_move,boardstate,xoro):
    """
    Format board based on moves
    player_move:
        description:
    
    boardstate:
        Description:
    
    xoro:
        Description:
    """
    form = '\n'+str(boardstate(player_move,xoro)[0])+ \
    '\n'+str(boardstate(player_move,xoro)[1]) + \
    '\n'+str(boardstate(player_move,xoro)[2])
    
    return (form)


def boardstate(loc,txto):
    """
    Representing the location chosen by Players (for either 'x' or 'o')
    The empty cells are replaced with either 'X' or 'O'
    
    attributes:
    loc:


    txto:

    """
    if loc == 1 and txto == 'x':
        board[0][0]=' X '
    elif loc == 2 and txto == 'x':
        board[0][1]=' X '
    elif loc == 3 and txto == 'x':
        board[0][2]=' X '
    elif loc == 4 and txto == 'x':
        board[1][0]=' X '
    elif loc == 5 and txto == 'x':
        board[1][1]=' X '
    elif loc == 6 and txto == 'x':
        board[1][2]=' X '
    elif loc == 7 and txto == 'x':
        board[2][0]=' X '
    elif loc == 8 and txto == 'x':
        board[2][1]=' X '
    elif loc == 9 and txto == 'x':
        board[2][2]=' X '
    if loc == 1 and txto == 'o':
        board[0][0]=' O '
    elif loc == 2 and txto == 'o':
        board[0][1]=' O '
    elif loc == 3 and txto == 'o':
        board[0][2]=' O '
    elif loc == 4 and txto == 'o':
        board[1][0]=' O '
    elif loc == 5 and txto == 'o':
        board[1][1]=' O '
    elif loc == 6 and txto == 'o':
        board[1][2]=' O '
    elif loc == 7 and txto == 'o':
        board[2][0]=' O '
    elif loc == 8 and txto == 'o':
        board[2][1]=' O '
    elif loc == 9 and txto == 'o':
        board[2][2]=' O '

    return (board)

def win_check():
"""
Defining the Win state
There are 8 states for each 'X' and 'O' where Win state is achieved
"""
    if (board[0][0] == ' X ' and board[0][1] == ' X ' and board[0][2] == ' X ') or \
       (board[1][0] == ' X ' and board[1][1] == ' X ' and board[1][2] == ' X ') or \
       (board[2][0] == ' X ' and board[2][1] == ' X ' and board[2][2] == ' X ') or \
       (board[0][0] == ' X ' and board[1][0] == ' X ' and board[2][0] == ' X ') or \
       (board[0][1] == ' X ' and board[1][1] == ' X ' and board[2][1] == ' X ') or \
       (board[0][2] == ' X ' and board[1][2] == ' X ' and board[2][2] == ' X ') or \
       (board[0][0] == ' X ' and board[1][1] == ' X ' and board[2][2] == ' X ') or \
       (board[0][2] == ' X ' and board[1][1] == ' X ' and board[2][0] == ' X ') or \
       (board[0][0] == ' O ' and board[0][1] == ' O ' and board[0][2] == ' O ') or \
       (board[1][0] == ' O ' and board[1][1] == ' O ' and board[1][2] == ' O ') or \
       (board[2][0] == ' O ' and board[2][1] == ' O ' and board[2][2] == ' O ') or \
       (board[0][0] == ' O ' and board[1][0] == ' O ' and board[2][0] == ' O ') or \
       (board[0][1] == ' O ' and board[1][1] == ' O ' and board[2][1] == ' O ') or \
       (board[0][2] == ' O ' and board[1][2] == ' O ' and board[2][2] == ' O ') or \
       (board[0][0] == ' O ' and board[1][1] == ' O ' and board[2][2] == ' O ') or \
       (board[0][2] == ' O ' and board[1][1] == ' O ' and board[2][0] == ' O '):
        return True
    else:
        return False

def draw_check():
"""
Defining state for Draw
No cells are empty
"""
    if (board[0][0] != empty_cell and board[0][1] != empty_cell and board[0][2] != empty_cell and \
        board[1][0] != empty_cell and board[1][1] != empty_cell and board[1][2] != empty_cell and \
        board[2][0] != empty_cell and board[2][1] != empty_cell and board[2][2] != empty_cell):
        return True
    else:
        return False

def P1_inputs(turn,board,xoro):   #can we breakdown this function ?
"""
Player input
Get available locations
Player selects move location
Check if selected location is empty
Check if the move leads to WIN
"""
    try:
        # Checking if turn is for Player 1 (X)
        if (turn % 2 == 0):

            player_move = int(input("please enter move (P-1)X:").strip())

            if player_move == 1:
                board_loc = board[0][0]
                print(board_loc)
            if player_move == 2:
                board_loc = board[0][1]
            if player_move == 3:
                board_loc = board[0][2]
            if player_move == 4:
                board_loc = board[1][0]
            if player_move == 5:
                board_loc = board[1][1]
            if player_move == 6:
                board_loc = board[1][2]
            if player_move == 7:
                board_loc = board[2][0]
            if player_move == 8:
                board_loc = board[2][1]
            if player_move == 9:
                board_loc = board[2][2]
#                 if player_move != [1-9]:
#                     print('Wrong location')

            if board_loc == empty_cell:
    #             boardstate(player_move,'x',0)
                
                print(player_move,boardstate,xoro)
                print(format_board(player_move,boardstate,xoro))
#                 cellno = current_available_cells()
#                 print('Available cells: ' + cellno)
                if win_check():
                    print('P-1 WINS')
                    board,turn = init_state_board()
                    # break
                    
                turn = turn + 1
        
                if draw_check():
                    print('GAME DRAWN')
                    board,turn = init_state_board()
                    # break
        
            else:
                print('Place already taken. Try again..')
#                 cellno = current_available_cells()
#                 print('Available cells: ' + cellno)
        else:
            player_move = int(input("please enter move (P-2)O:").strip())

            if player_move == 1:
                board_loc = board[0][0]
            if player_move == 2:
                board_loc = board[0][1]
            if player_move == 3:
                board_loc = board[0][2]
            if player_move == 4:
                board_loc = board[1][0]
            if player_move == 5:
                board_loc = board[1][1]
            if player_move == 6:
                board_loc = board[1][2]
            if player_move == 7:
                board_loc = board[2][0]
            if player_move == 8:
                board_loc = board[2][1]
            if player_move == 9:
                board_loc = board[2][2]
#                 if player_move != [1-9]:
#                     print('Wrong location')

            if board_loc == empty_cell:
    #             boardstate(player_move,'x',0)
                
                print(format_board(player_move,boardstate,xoro))
#                 cellno = current_available_cells()
#                 print('Available cells: ' + cellno)
                if win_check():
                    print('P-2 WINS')
                    board,turn = init_state_board()
                    # break
                    
                turn = turn + 1
        
                if draw_check():
                    print('GAME DRAWN')
                    board,turn = init_state_board()
                    # break
        
            else:
                print('Place already taken. Try again..')
#                 cellno = current_available_cells()
#                 print('Available cells: ' + cellno)

    except:
        print('Please enter a choice: ')
    return (board, turn)

# Driving program
new_game = 'y'
while new_game == 'y' or new_game == 'Y': # lower(new_game)
    new_game = str(input("Do you want to start a new game (y to play. press any other key to exit)?: ").strip())
    if new_game == 'y' or new_game == 'Y':
        print('Welcome to Tic-Tac-Toe')
        print('The inputs used are represented in the board as below')
        # New game starts here        
        # Reset the board to empty
        board,turn = init_state_board()
        while turn >= 0: 
            
            board, turn = P1_inputs(turn,board,'x')
            if turn == 0:
                break
#             print(turn)
            
            # Checking if turn is for Player 2 (O)
            board, turn = P1_inputs(turn,board,'o')
            if turn == 0:
                break
#             print(turn)
        else:
            break
    else:
        print('Thanks for playing')
```
